```python
import numpy as np
import pandas as pd
from collections import Counter

def summarize_behavior_time(experiment, include_none: bool = True):
    """
    Returns a DataFrame with total frames and seconds spent in each behavior
    across all trials in `experiment.trials`.
    
    experiment.trials should be a dict of {trial_name: SleapTrial instance}.
    Assumes each trial has:
      - trial.features_df with a 'behavior_active' column (NaN when none)
      - trial.frame_times array giving the timestamp of each row in features_df
    """
    # counters for frames and seconds
    frame_counts = Counter()
    sec_counts   = Counter()
    
    for name, trial in experiment.trials.items():
        # make sure behavior_active is there
        df = trial.features_df.copy()
        if 'behavior_active' not in df:
            # you can call: trial.add_behavior_column()
            raise RuntimeError(f"{name} missing behavior_active — run add_behavior_column() first")
        
        # fill NaN if you want to include “None”
        beh = df['behavior_active'].fillna('None') if include_none else df['behavior_active'].dropna()
        
        # count frames per behavior in this trial
        counts = beh.value_counts()
        
        # estimate dt from frame_times
        ft = trial.frame_times
        dt = np.mean(np.diff(ft)) if len(ft)>1 else 1.0/trial.fs
        
        for b, c in counts.items():
            frame_counts[b] += int(c)
            sec_counts[b]   += c * dt
    
    # build summary table
    summary = pd.DataFrame({
        'frames' : frame_counts,
        'seconds': sec_counts
    })
    summary.index.name = 'Behavior'
    # sort so “None” is last (optional)
    if include_none and 'None' in summary.index:
        none = summary.loc['None']
        summary = summary.drop('None')
        summary = pd.concat([summary, pd.DataFrame([none], index=['None'])])
    
    return summary

# --------------------------------------------
# Usage:
#  experiment = YourSleapExperiment(...)
# exp.extract_sleap_features()   # builds trial.features_df
for trial in exp.trials.values():
    trial.add_behavior_column()       # fills trial.features_df['behavior_active']

beh_summary = summarize_behavior_time(exp, include_none=True)
print(beh_summary)
```

```python
import pandas as pd
from collections import Counter, defaultdict

def summarize_aggression_defeat_with_trials(experiment):
    """
    For each mouse (subject_identity / intruder_identity), count how many frames
    of Aggression they performed and how many frames of Defeat they received,
    AND record which trial(s) each occurred in.
    
    Returns a DataFrame with columns:
      - mouse_identity
      - aggression_frames
      - aggression_trials    (list of trial names)
      - defeat_frames
      - defeat_trials        (list of trial names)
    """
    # Counters for frames
    agg_counts = Counter()
    def_counts = Counter()
    # Sets to collect trial names
    agg_trials = defaultdict(set)
    def_trials = defaultdict(set)
    
    # Loop over each trial
    for trial_name, trial in experiment.trials.items():
        df = trial.features_df
        if df is None or df.empty:
            continue
        
        # Aggression mask (performed by the subject)
        mask_agg = df["behavior_active"].str.contains("Aggression", na=False)
        # Defeat mask (received by the intruder)
        mask_def = df["behavior_active"].str.contains("Defeat",      na=False)
        
        # Count and record trials for aggression
        n_agg = int(mask_agg.sum())
        if n_agg > 0:
            subj = trial.subject_name
            agg_counts[subj] += n_agg
            agg_trials[subj].add(trial_name)
        
        # Count and record trials for defeat
        # Here we group by intruder_identity
        defeated = df.loc[mask_def, "intruder_identity"].dropna().tolist()
        for intr in defeated:
            def_counts[intr] += 1
            def_trials[intr].add(trial_name)
    
    # Build a DataFrame
    mice = sorted(set(agg_counts) | set(def_counts))
    rows = []
    for mouse in mice:
        rows.append({
            "mouse_identity"     : mouse,
            "aggression_frames"  : agg_counts.get(mouse, 0),
            "aggression_trials"  : sorted(agg_trials.get(mouse, [])),
            "defeat_frames"      : def_counts.get(mouse, 0),
            "defeat_trials"      : sorted(def_trials.get(mouse, [])),
        })
    
    summary = pd.DataFrame(rows)
    # If you like, sort by mouse_identity
    summary = summary.sort_values("mouse_identity").reset_index(drop=True)
    
    # Print it nicely
    print(summary.to_string(index=False))
    return summary

# Usage:
# exp = SleapExperiment(...)
# exp.extract_sleap_features()
# for tr in exp.trials.values():
#     tr.add_behavior_column()
summarize_aggression_defeat_with_trials(exp)
```

```python
import numpy as np
import pandas as pd
from collections import Counter

def summarize_social_masks(experiment, by_trial: bool = True, include_seconds: bool = True):
    """
    Summarize, across your SleapExperiment, how many frames (and optionally seconds)
    each trial spends with subject→agent or agent→subject social masks == "Yes".

    Parameters
    ----------
    experiment : SleapExperiment
        Your experiment, after extract_sleap_features() has been called.
    by_trial : bool
        If True, returns one row per trial; if False, aggregates over all trials.
    include_seconds : bool
        If True, also computes the approximate seconds = frames * dt.

    Returns
    -------
    pd.DataFrame
        Index = trial names (or a single row "ALL") with columns:
        - frames_total
        - frames_subject_in_agent
        - frames_agent_in_subject
        - seconds_total (if include_seconds)
        - seconds_subject_in_agent (if include_seconds)
        - seconds_agent_in_subject (if include_seconds)
    """
    records = []

    for name, trial in experiment.trials.items():
        # make sure you have both masks computed
        if not hasattr(trial, "mask_subject_in_agent") or not hasattr(trial, "mask_agent_in_subject"):
            # compute them on the full raw frame‐count
            trial.compute_social_ellipse_masks(a_cm=5, b_cm=3, use_full=True)

        df = trial.features_df

        # total frames in features_df
        n_frames = len(df)

        # count how many of those have each mask == "Yes"
        # (if your features_df stores them as strings "Yes"/"No")
        # otherwise just sum the boolean array
        n_subj_in_agent = (df["subject_in_agent"] == "Yes").sum()
        n_agent_in_subj = (df["agent_in_subject"] == "Yes").sum()

        # approximate dt (seconds per frame)
        dt = np.nan
        if include_seconds and hasattr(trial, "in_bout_times") and len(trial.in_bout_times) > 1:
            dt = np.mean(np.diff(trial.in_bout_times))

        rec = {
            "trial"                       : name,
            "frames_total"                : n_frames,
            "frames_subject_in_agent"     : int(n_subj_in_agent),
            "frames_agent_in_subject"     : int(n_agent_in_subj),
        }
        if include_seconds and not np.isnan(dt):
            rec["seconds_total"]                = n_frames * dt
            rec["seconds_subject_in_agent"]     = n_subj_in_agent * dt
            rec["seconds_agent_in_subject"]     = n_agent_in_subj * dt

        records.append(rec)

    df_summary = pd.DataFrame(records)

    if not by_trial:
        # aggregate into one row
        agg = {
            "frames_total": df_summary["frames_total"].sum(),
            "frames_subject_in_agent": df_summary["frames_subject_in_agent"].sum(),
            "frames_agent_in_subject": df_summary["frames_agent_in_subject"].sum(),
        }
        if include_seconds:
            agg["seconds_total"]            = df_summary["seconds_total"].sum()
            agg["seconds_subject_in_agent"]= df_summary["seconds_subject_in_agent"].sum()
            agg["seconds_agent_in_subject"]= df_summary["seconds_agent_in_subject"].sum()
        df_summary = pd.DataFrame([agg], index=["ALL"])
    else:
        df_summary = df_summary.set_index("trial")

    return df_summary


overall = summarize_social_masks(exp, by_trial=False)
print(overall)
```

```python
def summarize_social(df) -> dict:
    """
    Prints and returns counts and percentages of frames where
    subject→agent and agent→subject masks are “Yes”, plus total frames,
    and the combined total when either mask is “Yes”.
    """
    total = len(df)
    if total == 0:
        print("⚠️  No features to summarize (empty DataFrame).")
        return {}

    # individual counts
    subj2agent = (df["subject_in_agent"] == "Yes").sum()
    agent2subj = (df["agent_in_subject"] == "Yes").sum()

    # combined (union) count
    combined_mask = (df["subject_in_agent"] == "Yes") | (df["agent_in_subject"] == "Yes")
    combined    = combined_mask.sum()

    # percentages
    pct_s2a     = subj2agent / total * 100
    pct_a2s     = agent2subj / total * 100
    pct_comb    = combined    / total * 100

    summary = {
        "total_frames":             total,
        "subject→agent_frames":     subj2agent,
        "subject→agent_%":          pct_s2a,
        "agent→subject_frames":     agent2subj,
        "agent→subject_%":          pct_a2s,
        "either_direction_frames":  combined,
        "either_direction_%":       pct_comb,
    }

    # print it nicely
    print(f"Total frames:              {total}")
    print(f"Subject→Agent:             {subj2agent} frames ({pct_s2a:.1f}%)")
    print(f"Agent→Subject:             {agent2subj} frames ({pct_a2s:.1f}%)")
    print(f"Either direction combined: {combined} frames ({pct_comb:.1f}%)")

    return summary
```
